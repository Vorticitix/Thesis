Smoothed GSS for entire European Continent

```python
#if True use all adjecent grid boxes to calculate spatially averaged GSS over a total of 9 grid boxes
smoothing=True
from mpl_toolkits.basemap import Basemap
#Define relevant lead days
lead_dayz = [5]
#define coordinates of left and right, up and down edge of every box to read the file
lonz_0 = np.arange(-30,56.1,4,dtype=np.int64)
lonz_1 = np.arange(-26,60.1,4,dtype=np.int64)
latz_1 = np.arange(30,74.1,4,dtype=np.int64)
latz_0 = np.arange(34,78.1,4,dtype=np.int64)
#convert negative longitude coordinates to positive ones by adding 360 degrees
lonz_0 = [int(i+360) if i<0 else int(i) for i in lonz_0]
lonz_1 = [int(i+360) if i<=0 else int(i) for i in lonz_1]

#Season with month numbers
seasonz = {'MAM':[3,4,5],
          'JJA':[6,7,8],
          'SON':[9,10,11],
          'DJF':[12,1,2]}
#Define model
modelz = ['ERA5RF']

month_list = ['DJF','MAM','JJA','SON','DJF','MAM','JJA','SON']

extremez_list = ['persistent_hw','persistent_hw','persistent_hw','persistent_hw',
                'persistent_cw','persistent_cw','persistent_cw','persistent_cw']

extreme_labelz = ['Warm','Warm','Warm','Warm','Cold','Cold','Cold','Cold']
GSS_file = "Monthly_mean_GSS_lead_day_{}_lon_{}_{}_lat_{}_{}_{}.csv"
#Define color map
colorz = [(0,'#ff0000'),(0.25,'#ff6600'),(0.5,'#ffff00'),
          (0.7,'#33cc33'),(1,'#006600')]   
cmap = matplotlib.colors.LinearSegmentedColormap.from_list('custom', colorz, N=1024)

#Define coordinates to plot. These are 1 greater in dimension than the previois defined coordinates so plt.pcolormesh can use them properly
plot_lonz = np.arange(-30,60.1,4)
plot_latz = np.arange(30,78.1,4)

#Convert to 2D grid
plot_lonz_2D,plot_latz_2D = np.meshgrid(plot_lonz,plot_latz)
#Loop over different modelz
for model in modelz:
    #Loop over all seasons
    for i in range(len(month_list)):
        fig,axz=plt.subplots(2,1,figsize=(13.5,9))
        for j,lead_day in enumerate(lead_dayz):
            ax = axz[j]
            ax.set_title('Day {} Forecast'.format(lead_day))
            GSS_europe = np.empty((len(latz_0),len(lonz_1)))
#             GSS_europe.fill(np.nan)
            month = month_list[i]
            extreme = extremez_list[i]
            for x in range(len(latz_0)):
                for y in range(len(lonz_0)):
                    lat_0=latz_0[::-1][x];lat_1=latz_1[::-1][x]
                    lon_0=lonz_0[y];lon_1=lonz_1[y]
                    df = pd.read_csv(path+GSS_file.format(lead_day,lon_0,lon_1,lat_1,lat_0,model),index_col=0)
                    GSS = df.loc[month][extreme]
                    GSS_europe[x,y]=GSS
            if smoothing:
                GSS_europe = my_tools.average_2D_matrix(GSS_europe)
                np.save(path+'/figures/map_europe_GSS_{}_{}_day_{}_{}_smoothed.npy'.format(extreme,month,lead_day,model),
                   GSS_europe)

            m = Basemap(llcrnrlon=-148,llcrnrlat=30,urcrnrlon=60,urcrnrlat=78,
                         resolution='i', projection='cyl', lat_0 = 52.5, lon_0 = -60,ax=ax)

            m.drawmapboundary(fill_color='w')
#             m.fillcontinents(color='w',lake_color='w')
            m.drawcoastlines(linewidth=.5)
            m.drawcountries(linewidth=.5)
            parallels = np.arange(0.,81,10.)
            # labels = [left,right,top,bottom]
            m.drawparallels(parallels,labels=[False,True,True,False])
            meridians = np.arange(0,361.,20.)
            m.drawmeridians(meridians,labels=[True,False,False,True])
            im = m.pcolormesh(plot_lonz_2D,plot_latz_2D,GSS_europe,latlon='true',
                              cmap=cmap,vmin=0,vmax=1)
            print(GSS_europe)
        fig.subplots_adjust(left=0.025,right=0.975,bottom=0.2,top=0.9,wspace=0.175,hspace=0.3)
        cax = fig.add_axes([0.05, 0.1, 0.925, 0.05])
        fig.colorbar(im, cax=cax, orientation='horizontal',label='Gilbert Skill Score')
        fig.suptitle('Gilbert Skill Score {} Extreme {} {}'.format(extreme_labelz[i],month,model),
                    fontsize=18)
        sys.exit()
        if smoothing:
            fig.savefig(path+'/figures/map_europe_GSS_{}_day_3_5_{}_smoothed_1979.jpg'.format(extreme,month,model))

        else:
            fig.savefig(path+'/figures/map_europe_GSS_{}_{}_{}_1979.jpg'.format(extreme,month,model))
        plt.close(fig)
```

```python
import matplotlib.gridspec as gridspec
import matplotlib

coordinatez = [
(62,54,0,8),
(62,54,8,16),#Germany
(62,54,16,24),
(54,46,0,8), #Spain#Ukraine
(54,46,8,16), 
(54,46,16,24),
(46,38,0,8), #Spain
(46,38,8,16), #Ukraine #Spain
(46,38,16,24)] #Ukraine

seasonz = {'MAM':[3,4,5],
          'JJA':[6,7,8],
          'SON':[9,10,11],
          'DJF':[12,1,2]}

seasonz_list = ['DJF','MAM','JJA','SON',
               'DJF','MAM','JJA','SON']

extremez_list = ['persistent_hw','persistent_hw','persistent_hw','persistent_hw',
                'persistent_cw','persistent_cw','persistent_cw','persistent_cw']

extremez_labelz = ['Warm Extreme','Warm Extreme','Warm Extreme','Warm Extreme',
                'Cold Extreme','Cold Extreme','Cold Extreme','Cold Extreme']



colorz = [(0,'#ff0000'),(0.25,'#ff6600'),(0.5,'#ffff00'),
          (0.7,'#33cc33'),(1,'#006600')]   
cmap = matplotlib.colors.LinearSegmentedColormap.from_list('custom', colorz, N=1024)

lead_dayz = [1,3,5,7,9]
fcst_modelz = ['GFS','ERA5RF']
file = 'Monthly_mean_GSS_lead_day_{}_lon_{}_{}_lat_{}_{}_{}.csv'
for fcst_model in fcst_modelz:
    for lead_day in lead_dayz:
        fig,axz = plt.subplots(2,4,figsize=(16,11))
        fig.suptitle('Gilbert Skill Score per Season and per Region. Day {} Forecast {} '.format(lead_day,fcst_model))
        for i,ax in enumerate(axz.flat):

            ax.set_yticks([0,1/3,2/3,1])
            ax.set_xticks([0,1/3,2/3,1])
            ax.set_xticklabels(['0° E','8° E','16° E','24° E'])
            ax.set_yticklabels(['62° N','54° N','46° N','38° N'][::-1])
            inner_grid = gridspec.GridSpecFromSubplotSpec(3,3,subplot_spec=ax,
                                                             hspace=0,wspace=0)
            for j,inner in enumerate(inner_grid):
                lat_0,lat_1,lon_0,lon_1 = coordinatez[j]
                df = pd.read_csv(path+file.format(lead_day,lon_0,lon_1,
                                                  lat_1,lat_0,fcst_model),
                                index_col=0)
                GSS = df.loc[seasonz_list[i],extremez_list[i]]
#                 df.index = pd.to_datetime(df.index)
                ax1 = fig.add_subplot(inner)
                ax1.set_xticklabels([])
                ax1.set_yticklabels([])
                ax1.set_xticks([])
                ax1.set_yticks([])
                ax1.text(0.5,0.5,'{:.2f}'.format(GSS),
                         horizontalalignment='center',
                        verticalalignment='center',fontsize=15)
                ax1.set_facecolor(cmap(GSS))
            ax.set_title('{} {}'.format(extremez_labelz[i],seasonz_list[i]))
        fig.savefig(path+'figures/total_GSS_lead_day_{}_{}_per_event_type'.format(lead_day,fcst_model),dpi=150)
        plt.close(fig)
#             ax.set_aspect('equal')
```
