```python
%load_ext autoreload
%autoreload 2
%matplotlib qt
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
from glob import glob
#set path for my_tools script
import sys
sys.path.append('/home/onno/Thesis/Scripts')
import my_tools
from my_tools import file_dic, plot_dic
import matplotlib.ticker as plticker
import string
from statsmodels.stats.weightstats import DescrStatsW
from matplotlib.lines import Line2D

```

Line diagram mean RWP Errors all longitudes NH

```python
#set file paths
path_annual = '/home/onno/Thesis/GFS_mean/'
path_season = '/home/onno/Thesis/GFS_mean_season/'
path_to_write = '/home/onno/Thesis/Main_Figures/'

#set variables
variablez = ['envelope','phasespeed','u_wind']
y_labelz = ['E (m/s)','Cp (m/s)','u (m/s)']
#set ax titles
titlez = ['Annual','DJF','JJA']
letterz = np.array(list(string.ascii_lowercase)[:9]).reshape(3,3)
#deifne each lead hour
leads = np.linspace(0,240,11)
#set latitude data for NH and SH to plot
# loc = plticker.MultipleLocator(base=0.25,nbins='auto') 
# this locator puts ticks at regular intervals
latz = slice(70,40)
filez_annual = ['GFS_mean_error_{}_day_{:02d}_globe_mars.nc','ERA5RF_mean_error_{}_day_{:02d}_globe_mars.nc']
filez_DJF = ['GFS_mean_error_{}_DJF_day_{:02d}_globe.nc','ERA5RF_mean_error_{}_DJF_day_{:02d}.nc']
filez_JJA = ['GFS_mean_error_{}_JJA_day_{:02d}_globe.nc','ERA5RF_mean_error_{}_JJA_day_{:02d}.nc']
#intialize figure
fig,axz = plt.subplots(3,3,sharey='row',sharex=True,figsize=(10,7.5))
for k,variable in enumerate(variablez):
    #set relevant axis
    #create empty lists to add each dataset for each lead day for annual results, DJF and JJA
    

    #loop over lead days
    for j in np.arange(len(filez_annual)):
        datasets_annual = [];datasets_DJF = [];datasets_JJA = []
        for i in range(11):
            file_annual = filez_annual[j].format(variable,i)
            file_DJF = filez_DJF[j].format(variable,i)
            file_JJA = filez_JJA[j].format(variable,i)
            #open dataset for each lead day and append to empty lists. 
            #Also only select data between 75 and 25 degrees north and south
            try:
                ds_single = xr.open_dataset(path_annual+file_annual,decode_times=False).sel(lat=latz)
                datasets_annual.append(ds_single)
                ds_single_DJF = xr.open_dataset(path_season+file_DJF,decode_times=False).sel(lat=latz)
                datasets_DJF.append(ds_single_DJF)
                ds_single_JJA = xr.open_dataset(path_season+file_JJA,decode_times=False).sel(lat=latz)
                datasets_JJA.append(ds_single_JJA)
            except FileNotFoundError:
                file_annual = filez_annual[j].format(variable,5)
                file_DJF = filez_DJF[j].format(variable,5)
                file_JJA = filez_JJA[j].format(variable,5)
                ds_single = xr.open_dataset(path_annual+file_annual,decode_times=False).sel(lat=latz)
                datasets_annual.append(ds_single)
                ds_single_DJF = xr.open_dataset(path_season+file_DJF,decode_times=False).sel(lat=latz)
                datasets_DJF.append(ds_single_DJF)
                ds_single_JJA = xr.open_dataset(path_season+file_JJA,decode_times=False).sel(lat=latz)
                datasets_JJA.append(ds_single_JJA)
        #create 3D dataset with lead time as third dimension
        weights = np.cos(np.deg2rad(ds_single.lat))
        ds_annual = xr.concat(datasets_annual,dim='lead')
        ds_DJF = xr.concat(datasets_DJF,dim='lead')
        ds_JJA = xr.concat(datasets_JJA,dim='lead')
        var = list(ds_annual.keys())[0]
        
        array_annual_mean = DescrStatsW(ds_annual[var].mean('lon').values.T,weights=weights).mean
        array_DJF_mean = DescrStatsW(ds_DJF[var].mean('lon').values.T,weights=weights).mean
        array_JJA_mean = DescrStatsW(ds_JJA[var].mean('lon').values.T,weights=weights).mean
        
        array_annual_std = DescrStatsW(ds_annual[var].mean('lon').values.T,weights=weights).std
        array_DJF_std = DescrStatsW(ds_DJF[var].mean('lon').values.T,weights=weights).std
        array_JJA_std = DescrStatsW(ds_JJA[var].mean('lon').values.T,weights=weights).std
    
        axz[k,0].plot(array_annual_mean)
        axz[k,1].plot(array_DJF_mean)
        axz[k,2].plot(array_JJA_mean)
        axz[k,0].fill_between(np.arange(11),array_annual_mean-array_annual_std,
                              array_annual_mean+array_annual_std,alpha=0.25)
        axz[k,1].fill_between(np.arange(11),array_DJF_mean-array_DJF_std,
                      array_DJF_mean+array_DJF_std,alpha=0.25)
        axz[k,2].fill_between(np.arange(11),array_JJA_mean-array_JJA_std,
                      array_JJA_mean+array_JJA_std,alpha=0.25)
        
        axz[k,0].set_ylabel(y_labelz[k])

for ax in axz.flat:
    ax.set_xticks(np.arange(11))
    ax.grid(b=True,which='major',axis='y')
axz[0,0].set_title('Annual'),axz[0,1].set_title('DJF'),axz[0,2].set_title('JJA')
axz[2,0].set_xlabel('Lead Days'); axz[2,1].set_xlabel('Lead Days'); axz[2,2].set_xlabel('Lead Days')     
fig.suptitle('Forecast Biases Northern Hemisphere (40 °N - 70 °N)')

custom_lines = [Line2D([0], [0], color='tab:blue', lw=4),
                Line2D([0], [0], color='tab:orange', lw=4)]
axz[0,2].legend(custom_lines,['GEFSRF','ERA5RF'],loc='upper right', bbox_to_anchor=(1.03, 1.35))
fig.subplots_adjust(left=0.075,right=0.95,top=0.9,bottom=0.1,wspace=0.1,hspace=0.1) 
fig.savefig(path_to_write+'line_diagram_NH',dpi=200)
```
