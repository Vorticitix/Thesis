Combine all anomalies for all events into one csv dataframe

```python
lead_dayz = [3,5,7]
letterz = np.array(list(string.ascii_lowercase)[:9])
for lat_1,lat_0,lon_0,lon_1 in coordinatez:
    fig,axz = plt.subplots(3,3,figsize=(9,10))
    for i,lead_day in enumerate(lead_dayz):
        ax = axz[0,i]
        df_rean = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_pw_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_fcst = pd.read_csv(path+'/anom_files/fcst_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_pw_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median = df_rean.median(axis=0)
        df_fcst_median = df_fcst.median(axis=0)
        df_rean_q = df_rean.quantile([0.1,0.25,0.75,0.9])
        df_fcst_q = df_fcst.quantile([0.1,0.25,0.75,0.9])
        
        df_rean_50p = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_pw_50p_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median_50p = df_rean_50p.median(axis=0)
        df_rean_median_50p = df_rean_median - df_rean_median_50p
        
        df_MAE = pd.read_csv(path+'/anom_files/Mean_Absolute_Error_lon_{}_{}_lat_{}_{}_lead_day_{}_pw_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_MAE_median = df_MAE.median(axis=0)
        
        ax.plot(np.arange(1,11),df_rean_median,label='Reanalysis')
        ax.plot(np.arange(1,11),df_fcst_median,label='Forecast')

        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.75],df_rean_median,color='#1f77b4',alpha=.3)
        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.25],df_rean_median,color='#1f77b4',alpha=.3)
        ax.plot(np.arange(1,11),df_rean_q.loc[0.9],color='#1f77b4',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_rean_q.loc[0.1],color='#1f77b4',alpha=.5,linestyle='--')

        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.75],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.25],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.9],color='#ff7f0e',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.1],color='#ff7f0e',alpha=.5,linestyle='--')
        
        ax.plot(np.arange(1,11),df_rean_median_50p,color='r',label='50th percentile climatology')
        
        ax.plot(np.arange(1,11),df_MAE_median,color='g',label='MAE Forecast',linestyle='--')
        if i==1:
            ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.035),ncol=4)
        
        ax.set_xticks(np.arange(0,11))
        ax.margins(x=0)
        ax.grid()
        ax.set_xlabel('Lead Day')
        ax.set_ylabel('T Anomaly from 90th Perc. Clim. [째C] \n Mean Absolute Error [째C]')
        ax.set_ylim([-10,6])
        ax.set_title('Persistent Warm D-{}'.format(lead_day))
        ax.axhline(y=0,color='k')
        ax.axvline(x=lead_day,color='k')
        ax.label_outer()
#         fig.savefig(path+'/anom_files/pw_lon_{}_{}_lat_{}_{}_lead_day_{}.png'.format(lon_0,lon_1,lat_1,lat_0,lead_day))
#         plt.close(fig)

    for i,lead_day in enumerate(lead_dayz):
        ax = axz[1,i]
        df_rean = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_cw_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_fcst = pd.read_csv(path+'/anom_files/fcst_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_cw_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median = df_rean.median(axis=0)
        df_fcst_median = df_fcst.median(axis=0)
        df_rean_q = df_rean.quantile([0.1,0.25,0.75,0.9])
        df_fcst_q = df_fcst.quantile([0.1,0.25,0.75,0.9])
        
        df_rean_50p = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_cw_50p_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median_50p = df_rean_50p.median(axis=0)
        df_rean_median_50p = df_rean_median - df_rean_median_50p
        
        df_MAE = pd.read_csv(path+'/anom_files/Mean_Absolute_Error_lon_{}_{}_lat_{}_{}_lead_day_{}_pc_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_MAE_median = df_MAE.median(axis=0)
        
        ax.plot(np.arange(1,11),df_rean_median,label='Reanalysis')
        ax.plot(np.arange(1,11),df_fcst_median,label='Forecast')

        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.75],df_rean_median,color='#1f77b4',alpha=.3)
        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.25],df_rean_median,color='#1f77b4',alpha=.3)
        ax.plot(np.arange(1,11),df_rean_q.loc[0.9],color='#1f77b4',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_rean_q.loc[0.1],color='#1f77b4',alpha=.5,linestyle='--')

        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.75],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.25],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.9],color='#ff7f0e',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.1],color='#ff7f0e',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_rean_median_50p,color='r',label='50th percentile climatology')
        
        ax.plot(np.arange(1,11),df_MAE_median,color='g',label='MAE Forecast',linestyle='--')
        
        if i==1:
            ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.035),ncol=4)        
        ax.set_xticks(np.arange(0,11))
        ax.margins(x=0)
        ax.grid()
        ax.set_xlabel('Lead Day')
        ax.set_ylabel('T Anomaly from 10th Perc. Clim. [째C] \n Mean Absolute Error [째C]')
        ax.set_title('Persistent Cold D-{}'.format(lead_day))
        ax.set_ylim([-6,10])
        ax.axhline(y=0,color='k')
        ax.axvline(x=lead_day,color='k')
        ax.label_outer()

    for i,lead_day in enumerate(lead_dayz):
        ax = axz[2,i]
        df_rean = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_average_50p_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_fcst = pd.read_csv(path+'/anom_files/fcst_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_average_50p_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median = df_rean.median(axis=0)
        df_fcst_median = df_fcst.median(axis=0)
        df_rean_q = df_rean.quantile([0.1,0.25,0.75,0.9])
        df_fcst_q = df_fcst.quantile([0.1,0.25,0.75,0.9])
        
        df_rean_60p = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_average_60p_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median_60p = df_rean_60p.median(axis=0)
        df_rean_median_60p = df_rean_median - df_rean_median_60p
        
        df_rean_40p = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_average_40p_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median_40p = df_rean_40p.median(axis=0)
        df_rean_median_40p = df_rean_median - df_rean_median_40p
        
        df_MAE = pd.read_csv(path+'/anom_files/Mean_Absolute_Error_lon_{}_{}_lat_{}_{}_lead_day_{}_avg_new.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_MAE_median = df_MAE.median(axis=0)
        
        ax.plot(np.arange(1,11),df_rean_median,label='Reanalysis')
        ax.plot(np.arange(1,11),df_fcst_median,label='Forecast')

        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.75],df_rean_median,color='#1f77b4',alpha=.3)
        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.25],df_rean_median,color='#1f77b4',alpha=.3)
        ax.plot(np.arange(1,11),df_rean_q.loc[0.9],color='#1f77b4',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_rean_q.loc[0.1],color='#1f77b4',alpha=.5,linestyle='--')

        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.75],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.25],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.9],color='#ff7f0e',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.1],color='#ff7f0e',alpha=.5,linestyle='--')
        
        ax.plot(np.arange(1,11),df_rean_median_60p,color='r',label='40th/60th percentile climatology')
        ax.plot(np.arange(1,11),df_rean_median_40p,color='r')
        
        ax.plot(np.arange(1,11),df_MAE_median,color='g',label='MAE Forecast',linestyle='-.')
        
        ax.set_xticks(np.arange(0,11))
        ax.margins(x=0)
        ax.grid()
        ax.set_xlabel('Lead Day')
        ax.set_ylabel('T Anomaly from 50th Perc. Clim. [째C] \n Mean Absolute Error [째C]')
        ax.set_title('Persistent Average D-{}'.format(lead_day))
        ax.set_ylim([-8,8])
        if i==1:
            ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.175),ncol=4)       
        ax.axhline(y=0,color='k')
        ax.axvline(x=lead_day,color='k')
        ax.label_outer()
    for j,ax in enumerate(axz.flat):
        ax.text(x=0,y=1.09,s=letterz[j]+')',transform=ax.transAxes,fontweight='bold',fontsize=13,
                       verticalalignment='top')
        
#     fig.suptitle('Persistent Avergae Weather Temperature Anomalies in Forecast and Reanalysis {}-Day Forecasts\n{}-{} 째N {}-{} 째E'\
#         .format(lead_day,lat_0,lat_1,lon_0,lon_1,df_rean.shape[0]))
    fig.subplots_adjust(left=0.1,right=0.98,bottom=0.1,top=0.95,wspace=0.1,hspace=0.3)
    fig.savefig('/media/onno/Volume/Main_Figures/anom_plots/all_lon_{}_{}_lat_{}_{}.png'.format(lon_0,lon_1,lat_1,lat_0,lead_day))
    plt.close(fig)
```
