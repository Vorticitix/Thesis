Plot persistent warm extremes

```python
lead_dayz = [3,5,7]
for lat_1,lat_0,lon_0,lon_1 in coordinatez:
    fig,axz = plt.subplots(1,3,figsize=(9,5))
    for i,lead_day in enumerate(lead_dayz):
        ax = axz.flat[i]
        df_rean = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_pw.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_fcst = pd.read_csv(path+'/anom_files/fcst_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_pw.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median = df_rean.median(axis=0)
        df_fcst_median = df_fcst.median(axis=0)
        df_rean_q = df_rean.quantile([0.1,0.25,0.75,0.9])
        df_fcst_q = df_fcst.quantile([0.1,0.25,0.75,0.9])
        
        df_rean_50p = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_pw_50p.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_median_50p = df_rean_50p.median(axis=0)
        df_rean_median_50p = df_rean_median - df_rean_median_50p
        
        df_MAE = pd.read_csv(path+'/anom_files/Mean_Absolute_Error_lon_{}_{}_lat_{}_{}_lead_day_{}_pw.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_MAE_median = df_MAE.median(axis=0)
        
        ax.plot(np.arange(1,11),df_rean_median,label='Reanalysis')
        ax.plot(np.arange(1,11),df_fcst_median,label='Forecast')

        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.75],df_rean_median,color='#1f77b4',alpha=.3)
        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.25],df_rean_median,color='#1f77b4',alpha=.3)
        ax.plot(np.arange(1,11),df_rean_q.loc[0.9],color='#1f77b4',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_rean_q.loc[0.1],color='#1f77b4',alpha=.5,linestyle='--')

        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.75],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.25],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.9],color='#ff7f0e',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.1],color='#ff7f0e',alpha=.5,linestyle='--')
        
        ax.plot(np.arange(1,11),df_rean_median_50p,color='r',label='50th percentile climatology')
        
        ax.plot(np.arange(1,11),df_MAE_median,color='g',label='MAE Forecast',linestyle='--')
        if i==1:
            ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),ncol=4)
        
        ax.set_xticks(np.arange(0,11))
        ax.margins(x=0)
        ax.grid()
        ax.set_xlabel('Lead Day')
        ax.set_ylabel('Temperature Anomaly from 90th Percentile Climatology [°C] \n Mean Abolute Error [°C]')
        ax.set_ylim([-10,6])
        ax.set_title('Lead Day {}'.format(lead_day))
        ax.axhline(y=0,color='k')
        ax.axvline(x=lead_day,color='k')
        ax.label_outer()
    fig.suptitle('Persistent Warm Extremes Temperature Anomalies in Forecast and Reanalysis {}-Day Forecasts\n{}-{} °N {}-{} °E'\
        .format(lead_day,lat_0,lat_1,lon_0,lon_1,df_rean.shape[0]))
    fig.subplots_adjust(left=0.1,right=0.98,bottom=0.2,top=0.85,wspace=0.075,hspace=0.05)
    fig.savefig('/media/onno/Volume/Main_Figures/anom_plots/pw_lon_{}_{}_lat_{}_{}.png'.format(lon_0,lon_1,lat_1,lat_0,lead_day))
    plt.close(fig)
#         fig.savefig(path+'/anom_files/pw_lon_{}_{}_lat_{}_{}_lead_day_{}.png'.format(lon_0,lon_1,lat_1,lat_0,lead_day))
#         plt.close(fig)
```

```python
lead_dayz = [3,5,7]
for lead_day in lead_dayz:
    for lat_1,lat_0,lon_0,lon_1 in coordinatez:
        fig,ax = plt.subplots(figsize=(12,9))
        df_rean = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_average_50p.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_fcst = pd.read_csv(path+'/anom_files/fcst_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_average_50p.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean.index = pd.to_datetime(df_rean.index); df_fcst.index = pd.to_datetime(df_fcst.index)
        df_rean = df_rean.loc[(df_rean.index.month>=12)|(df_rean.index.month<=2)]
        df_fcst = df_fcst.loc[(df_fcst.index.month>=12)|(df_fcst.index.month<=2)]
        
        df_rean_median = df_rean.median(axis=0)
        df_fcst_median = df_fcst.median(axis=0)
        df_rean_q = df_rean.quantile([0.1,0.25,0.75,0.9])
        df_fcst_q = df_fcst.quantile([0.1,0.25,0.75,0.9])
        
        df_rean_60p = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_average_60p.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_60p.index = pd.to_datetime(df_rean_60p.index)
        df_rean_60p = df_rean_60p.loc[(df_rean_60p.index.month>=12)|(df_rean_60p.index.month<=2)]
        df_rean_median_60p = df_rean_60p.median(axis=0)
        df_rean_median_60p = df_rean_median - df_rean_median_60p
        
        df_rean_40p = pd.read_csv(path+'/anom_files/rean_lon_{}_{}_lat_{}_{}_anoms_lead_day_{}_average_40p.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_rean_40p.index = pd.to_datetime(df_rean_40p.index)
        df_rean_40p = df_rean_40p.loc[(df_rean_40p.index.month>=12)|(df_rean_40p.index.month<=2)]
        df_rean_median_40p = df_rean_40p.median(axis=0)
        df_rean_median_40p = df_rean_median - df_rean_median_40p
        
        df_MAE = pd.read_csv(path+'/anom_files/Mean_Absolute_Error_lon_{}_{}_lat_{}_{}_lead_day_{}_avg.csv'.format(lon_0,lon_1,lat_1,lat_0,lead_day),index_col=0)
        df_MAE.index = pd.to_datetime(df_MAE.index)
        df_MAE = df_MAE.loc[(df_MAE.index.month>=12)|(df_MAE.index.month<=2)]
        df_MAE_median = df_MAE.median(axis=0)
        
        ax.plot(np.arange(1,11),df_rean_median,label='Reanalysis')
        ax.plot(np.arange(1,11),df_fcst_median,label='Forecast')

        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.75],df_rean_median,color='#1f77b4',alpha=.3)
        ax.fill_between(np.arange(1,11),df_rean_q.loc[0.25],df_rean_median,color='#1f77b4',alpha=.3)
        ax.plot(np.arange(1,11),df_rean_q.loc[0.9],color='#1f77b4',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_rean_q.loc[0.1],color='#1f77b4',alpha=.5,linestyle='--')

        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.75],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.fill_between(np.arange(1,11),df_fcst_q.loc[0.25],df_fcst_median,color='#ff7f0e',alpha=.3)
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.9],color='#ff7f0e',alpha=.5,linestyle='--')
        ax.plot(np.arange(1,11),df_fcst_q.loc[0.1],color='#ff7f0e',alpha=.5,linestyle='--')
        
        ax.plot(np.arange(1,11),df_rean_median_60p,color='r',label='40th/60th percentile climatology')
        ax.plot(np.arange(1,11),df_rean_median_40p,color='r')
        
        ax.plot(np.arange(1,11),df_MAE_median,color='g',label='MAE Forecast',linestyle='-.')
        
        ax.legend(loc='upper right')
        ax.set_xticks(np.arange(0,11))
        ax.margins(x=0)
        ax.grid()
        ax.set_xlabel('Lead Day')
        ax.set_ylabel('Anomaly from 50th percentile climatology [°C]')
        ax.set_ylim([-8,8])
        ax.axhline(y=0,color='k')
        ax.axvline(x=lead_day,color='k')
        fig.suptitle('Persistent Average Weather Temperature Anomalies in Forecast and Reanalysis {}-Day Forecasts {}-{} °N {}-{} °E : N = {}'\
        .format(lead_day,lat_0,lat_1,lon_0,lon_1,df_rean.shape[0]))
        fig.savefig(path+'/anom_files/average_lon_{}_{}_lat_{}_{}_lead_day_{}_DJF.png'.format(lon_0,lon_1,lat_1,lat_0,lead_day))
        plt.close(fig)
```
