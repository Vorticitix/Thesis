```python
%load_ext autoreload
%autoreload 2
import xarray as xr
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D
import cartopy.crs as ccrs
#set directory for importinh my_tools,py
import sys
sys.path.append('/home/onno/Thesis/Scripts')
import my_tools
from my_tools import file_dic, plot_dic
from datetime import datetime
import glob
from cartopy.util import add_cyclic_point
from mpl_toolkits.basemap import Basemap, addcyclic
import matplotlib

%matplotlib qt
```

```python
path_gfs = '/home/onno//Thesis/GFS_climatology/'
path_era = '/home/onno//Thesis/ERA_climatology/'
```

```python
variablez = ['envelope','phasespeed','u_wind']
custom_lines = [Line2D([0], [0], color='k', lw=4),
                Line2D([0], [0], color='r', lw=4)]
latz_hem = [90,-90]
levelz_contourf = [np.linspace(10,26,17),np.linspace(0,12,13),np.linspace(10,40,31)]
levelz_contour = [np.linspace(10,25,4),np.linspace(0,12,4),np.linspace(10,40,7)]
for j,variable in enumerate(variablez):
    era = xr.open_dataset(path_era+'ERA5_{}_clim_mars.nc'.format(variable))
    for lat_hem in latz_hem:
        for i in range(11):
            gfs = xr.open_dataset(path_gfs+'GFS_{}_clim_day_{:02d}.nc'.format(variable,i),decode_times=False)
            gfs_cyclic, lon_cyclic = addcyclic(gfs[plot_dic[variable]['data_variable']].values, gfs.lon)
            era_cyclic, lon_cyclic = addcyclic(era[plot_dic[variable]['data_variable']].values, era.lon)
            lon,lat = np.meshgrid(lon_cyclic,gfs.lat)
            fig,ax = plt.subplots(figsize=(16,9))
            m = Basemap(projection='ortho',lon_0=0,lat_0=lat_hem,resolution='c',ax=ax)
            cl1 = m.contour(lon,lat,era_cyclic,colors='k',levels=levelz_contour[j],latlon='true')
            cl2 = m.contour(lon,lat,gfs_cyclic,colors='r',levels=levelz_contour[j],latlon='true',linestyles='dashed')
            ax.clabel(cl2, fmt='%2.0f', colors='r', fontsize=12)
            ax.legend(custom_lines,['ERA5','GFS'])
            m.drawcoastlines()
            m.drawparallels((-80,-60,-40,-20,0,20,40,60,80))
            m.drawmeridians((0,60,120,180,240,300),labels=[0,0,1,1])
            fig.suptitle('GEFS {} Climatology vs ERA {} Climatology Day {}'.format(plot_dic[variable]['title'],plot_dic[variable]['title'],i))
            if lat_hem==90:
                fig.savefig(path_gfs+'/GFS_vs_ERA/GFS_vs_ERA_climatology_{}_day_{:02d}_NH'.format(variable,i))
            else:
                fig.savefig(path_gfs+'/GFS_vs_ERA/GFS_vs_ERA_climatology_{}_day_{:02d}_SH'.format(variable,i))
            plt.close(fig)

    
```
