```python
#set path for forecast data
path = '/media/onno/Algemeen/Thesis/GFS_T850/rank_forecast/'
#set path for opening general files
path_gen = '/media/onno/Algemeen/Thesis/'
```

```python
file_GFS_env = path_gen + file_dic['envelope']['GFS']
file_ERA5RF_env = path_gen + file_dic['envelope']['ERA5RF']
file_GFS_cp = path_gen + file_dic['phasespeed']['GFS']
file_ERA5RF_cp = path_gen + file_dic['phasespeed']['ERA5RF']
file_rean_env = path_gen + file_dic['envelope']['ERA5']
file_rean_cp = path_gen + file_dic['phasespeed']['ERA5']

GFS_env = xr.open_dataset(file_GFS_env, decode_times=False).squeeze()
ERA5RF_env = xr.open_dataset(file_ERA5RF_env, decode_times=False).squeeze()
GFS_cp = xr.open_dataset(file_GFS_cp,decode_times=False).squeeze()
ERA5RF_cp = xr.open_dataset(file_ERA5RF_cp,decode_times=False).squeeze()
rean_env = xr.open_dataset(file_rean_env).squeeze()
rean_cp = xr.open_dataset(file_rean_cp).squeeze()

GFS_init_time = pd.Timestamp('1800-01-01')
ERA5RF_init_time = pd.Timestamp('1900-01-01')
GFS_env['time']=[pd.Timedelta(i,'hours')+GFS_init_time for i in GFS_env.time.values]
GFS_cp['time']=[pd.Timedelta(i,'hours')+GFS_init_time for i in GFS_cp.time.values]
ERA5RF_env['time']=[pd.Timedelta(i,'hours')+ERA5RF_init_time for i in ERA5RF_env.time.values]
ERA5RF_cp['time']=[pd.Timedelta(i,'hours')+ERA5RF_init_time for i in ERA5RF_cp.time.values]

GFS_env = GFS_env.rename({'lon':'longitude','lat':'latitude'}); GFS_cp = GFS_cp.rename({'lon':'longitude','lat':'latitude'})
ERA5RF_env = ERA5RF_env.rename({'lon':'longitude','lat':'latitude'}); ERA5RF_cp = ERA5RF_cp.rename({'lon':'longitude','lat':'latitude'})
GFS_env = GFS_env.assign_coords(longitude=(((GFS_env.longitude + 180) % 360) - 180)).sortby('longitude').sel(latitude=slice(90,0))
GFS_cp = GFS_cp.assign_coords(longitude=(((GFS_cp.longitude + 180) % 360) - 180)).sortby('longitude').sel(latitude=slice(90,0))
ERA5RF_env = ERA5RF_env.assign_coords(longitude=(((ERA5RF_env.longitude + 180) % 360) - 180)).sortby('longitude').sel(latitude=slice(90,0))
ERA5RF_cp = ERA5RF_cp.assign_coords(longitude=(((ERA5RF_cp.longitude + 180) % 360) - 180)).sortby('longitude').sel(latitude=slice(90,0))
# GFS_env = GFS_env.sel(lead=GFS_env.lead.values[[bool(i) for i in GFS_env.lead%24==0]])
# GFS_cp = GFS_cp.sel(lead=GFS_cp.lead.values[[bool(i) for i in GFS_cp.lead%24==0]])
# ERA5RF_env = ERA5RF_env.sel(lead=ERA5RF_env.lead.values[[bool(i) for i in ERA5RF_env.lead%24==0]])
# ERA5RF_cp = ERA5RF_cp.sel(lead=ERA5RF_cp.lead.values[[bool(i) for i in ERA5RF_cp.lead%24==0]])

# GFS_env['lead'] = GFS_env.lead.values//24
# GFS_cp['lead'] = GFS_cp.lead.values//24
# ERA5RF_env['lead'] = ERA5RF_env.lead.values//24
# ERA5RF_cp['lead'] = ERA5RF_cp.lead.values//24


rean_env = rean_env.assign_coords(longitude=(((rean_env.longitude + 180) % 360) - 180)).sortby('longitude')
rean_env = rean_env.sel(latitude=slice(90,0))
rean_cp = rean_cp.assign_coords(longitude=(((rean_cp.longitude + 180) % 360) - 180)).sortby('longitude')
rean_cp = rean_cp.sel(latitude=slice(90,0))

```

```python
#list of seasons to loop over
seasonz = ['DJF','MAM','JJA','SON']
#list of events to loop over
eventz = ['persistent_hw','persistent_cw','short_hw','short_cw']
event_titlez = ['Persistent Heatwave','Persistent Coldwave','Short Heatwave','Short Coldwave']

rankz = ['good','bad']
```

RWP Envelope

```python
lat_0,lat_1,lon_0,lon_1 = (54,46,6,14)

file = '{}_forecasts_{}_lon_{}_{}_lat_{}_{}_{}_{}_lead_day_{}.txt'
eventz = ['persistent_hw']
seasonz = ['JJA']
modelz = ['ERA5RF']
fcstz = [ERA5RF_env]
lead_dayz = [3,5]

for j,model in enumerate(modelz):
    fcst_env = fcstz[j]
    for i,event in enumerate(eventz):
        datez_event = np.load(path_gen + 'GFS_T850/{}_lon_{}_{}_lat_{}_{}.npy'.format(event,lon_0,lon_1-2,lat_0-2,lat_1))
        for season in seasonz:
            for lead_day in lead_dayz:
                good_sub = file.format('good',event,lon_0,lon_1,lat_0,lat_1,season,model,lead_day)
                bad_sub = file.format('bad',event,lon_0,lon_1,lat_0,lat_1,season,model,lead_day)
                df_good = pd.read_csv(path+good_sub,index_col=0)
                df_good.index = pd.to_datetime(df_good.index)
                df_bad = pd.read_csv(path+bad_sub,index_col=0)
                df_bad.index = pd.to_datetime(df_bad.index)
                all_meanz = []
                all_meanz = []
                dfz = [df_good,df_bad]
                fig,axz = plt.subplots(1,2,figsize=(16,9))
                for df in dfz:
                    if df.equals(df_good):
                        ax = axz.flat[0]
                        title='Good N = {}' 
                    else:
                        ax = axz.flat[1]
                        title = 'Bad N = {}'
                    for begin_date_event in df.index:
                        idx = np.squeeze(np.where(begin_date_event==datez_event[:,0]))
                        end_date_event = pd.Timestamp(datez_event[idx,1])
                        begin_date_fcst = begin_date_event - pd.Timedelta(lead_day*24,'hours')
                        end_date_fcst = begin_date_fcst + pd.Timedelta(9*24,'hours')

                        if end_date_event>end_date_fcst:
                            end_date_event = end_date_fcst
                        end_lead_hours_object = ((end_date_event - begin_date_fcst) + pd.Timedelta(18,'h'))
                        end_lead_hours = end_lead_hours_object.days*24 + end_lead_hours_object.seconds//3600 

                        rean_env_sub = rean_env.sel(time=slice(begin_date_event,
                                                      end_date_event+pd.Timedelta(18,'hours')))
                        if model=='ERA5RF':
                            rean_env_sub = rean_env_sub.sel(time = rean_env_sub.time.values[[bool(i) for i in rean_env_sub.time.dt.hour%12==0]])
                        fcst_env_sub = fcst_env.sel(time=begin_date_fcst,lead=slice(lead_day*24,end_lead_hours))
                        diff_fcst = fcst_env_sub.v - rean_env_sub.v.values
                        all_meanz.append(diff_fcst.mean(dim='lead'))


                    diff_env_sub_all = xr.concat(all_meanz,dim='time').mean(dim='time').sel(latitude=slice(85,0))

                    diff_env_sub_cyclic, lon_cyclic = addcyclic(diff_env_sub_all.values, diff_env_sub_all.longitude)
                    m = Basemap(projection='ortho', 
                          lat_0=90, lon_0=0,ax=ax)

                    m.drawcoastlines()
                    latz,lonz = np.meshgrid(diff_env_sub_all.latitude,lon_cyclic)
                    im = m.contourf(lonz,latz,diff_env_sub_cyclic.T,
                               latlon='true',cmap='bwr',levels=np.arange(-10,10.5,.5),extend='both')
                    ax.set_title(title.format(len(df)))
                fig.suptitle('RWP Envelope Composite Forecast Error Forecasts Germany {} {} {} Day {} Forecast'.format(
                                                                                                 event_titlez[i],
                                                                                                season, model,lead_day))
                cax = fig.add_axes([0.2, 0.1, 0.6, 0.03])
                cbar = fig.colorbar(im,orientation='horizontal',cax=cax)
                cbar.ax.set_xlabel('Envelope Forecast Error (m/s)')
                fig.savefig(path_gen+'/fcst_composite/diff_fcst_rean/RWP_Envelope_diff_fcst_rean_Composite_{}_{}_{}_lead_day_{}'.format(model,event,season, lead_day),dpi=200)
                plt.close(fig)

```
