```python
%load_ext autoreload
%autoreload 2
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import sys
sys.path.append('/home/onno/Thesis/Scripts')
import my_tools
from cmap import ncl_colormap 
from mpl_toolkits.basemap import Basemap, addcyclic
from my_tools import file_dic, plot_dic
%matplotlib qt
```

```python
#set path for forecast data
path = '/media/onno/Algemeen/Thesis/GFS_T850/rank_forecast/seasonal/'
#set path for opening general files
path_gen = '/media/onno/Algemeen/Thesis/'
```

```python

#Open 500 hPa geopotential height data
file_rean_env = path_gen + file_dic['envelope']['ERA5']
file_rean_cp = path_gen + file_dic['phasespeed']['ERA5']
file_rean_z500 = path_gen + file_dic['Z500']['ERA5']
file_rean_mslp = path_gen + file_dic['MSLP']['ERA5']
rean_env_gen = xr.open_dataset(file_rean_env).squeeze()
rean_env_gen = rean_env_gen.assign_coords(longitude=(((rean_env_gen.longitude + 180) % 360) - 180)).sortby('longitude')
rean_env = rean_env_gen.sel(latitude=slice(90,0),
               time=[bool(i) for i in rean_env_gen.time.dt.hour%24==0])
rean_cp_gen = xr.open_dataset(file_rean_cp).squeeze()
rean_cp_gen = rean_cp_gen.assign_coords(longitude=(((rean_cp_gen.longitude + 180) % 360) - 180)).sortby('longitude')
rean_cp = rean_cp_gen.sel(latitude=slice(90,0),
               time=[bool(i) for i in rean_cp_gen.time.dt.hour%24==0])
rean_z500_gen = xr.open_dataset(file_rean_z500).squeeze()
rean_z500_gen = rean_z500_gen.assign_coords(longitude=(((rean_z500_gen.longitude + 180) % 360) - 180)).sortby('longitude')
rean_z500 = rean_z500_gen.sel(latitude=slice(90,0),
               time=[bool(i) for i in rean_z500_gen.time.dt.hour%24==0])
rean_mslp_gen = xr.open_dataset(file_rean_mslp).squeeze()
rean_mslp_gen = rean_mslp_gen.rename({'lon':'longitude','lat':'latitude'})
rean_mslp_gen = rean_mslp_gen.assign_coords(longitude=(((rean_mslp_gen.longitude + 180) % 360) - 180)).sortby('longitude')
rean_mslp = rean_mslp_gen.sel(latitude=slice(90,0),
               time=[bool(i) for i in rean_mslp_gen.time.dt.hour%24==0])

```

```python
#list of seasons to loop over
seasonz = ['DJF','MAM','JJA','SON']
#list of events to loop over
eventz = ['persistent_hw','persistent_cw','short_hw','short_cw']
event_titlez = ['Persistent Heatwave','Persistent Coldwave','Short Heatwave','Short Coldwave']

rankz = ['good','bad']

```

Seasonal Composite Analysis RWP Envelope

```python
lat_0,lat_1,lon_0,lon_1 = (54,46,6,14)
file = '{}_forecasts_{}_lon_{}_{}_lat_{}_{}_{}_{}.txt'
seasonz = ['DJF','JJA']
modelz = ['GFS','ERA5RF']

for model in modelz:
    for i,event in enumerate(eventz):
        datez_event = np.load(path_gen + 'GFS_T850/{}_lon_{}_{}_lat_{}_{}.npy'.format(event,lon_0,lon_1-2,lat_0-2,lat_1))
        for season in seasonz:
            for rank in rankz:
                file_sub = file.format(rank,event,lon_0,lon_1,lat_0,lat_1,season,model)
                df = pd.read_csv(path+file_sub,index_col=0)
                df.index = pd.to_datetime(df.index)
                all_meanz = []
                for date in df.index:
                    index = int(np.squeeze(np.where(datez_event[:,0]==date)))
                    date_event = datez_event[index]
                    rean_env_sub = rean_env.sel(time=slice(date_event[0],
                                                  date_event[1])).mean(dim='time')
                    all_meanz.append(rean_env_sub)

                fig,ax = plt.subplots(figsize=(16,9))
                rean_env_sub_all = xr.concat(all_meanz,dim='time').mean(dim='time')
                rean_env_sub_cyclic, lon_cyclic = addcyclic(rean_env_sub_all.v.values, rean_env_sub.longitude)
                m = Basemap(projection='ortho', 
                      lat_0=90, lon_0=0)

        #                 ax=ax)
                m.drawcoastlines()
                latz,lonz = np.meshgrid(rean_env_sub.latitude,lon_cyclic)
                im = m.contourf(lonz,latz,rean_env_sub_cyclic.T,
                           latlon='true',cmap=ncl_colormap(),levels=np.arange(41),extend='both')
                fig.suptitle('RWP Envelope Composite Analysis {} Forecasts Germany {} {} {}'.format(rank.title(),
                                                                                                 event_titlez[i],
                                                                                                season, model))
                cax = fig.add_axes([0.9, 0.1, 0.05, 0.8])
                ax.text(0.1,0.5,s='N = {}'.format(len(df.index)),transform=fig.transFigure,fontsize=16)
                fig.colorbar(im,orientation='vertical',cax=cax)
                fig.savefig(path_gen+'/fcst_composite/RWP_Envelope_Composite_{}_{}_{}_{}'.format(model,event,season,rank),dpi=200)
                plt.close(fig)

            
```

```python
lat_0,lat_1,lon_0,lon_1 = (54,46,6,14)
file = '{}_forecasts_{}_lon_{}_{}_lat_{}_{}_{}_{}.txt'
seasonz = ['DJF','JJA']
modelz = ['GFS','ERA5RF']

for model in modelz:
    for i,event in enumerate(eventz):
        datez_event = np.load(path_gen + 'GFS_T850/{}_lon_{}_{}_lat_{}_{}.npy'.format(event,lon_0,lon_1-2,lat_0-2,lat_1))
        for season in seasonz:
            for rank in rankz:
                file_sub = file.format(rank,event,lon_0,lon_1,lat_0,lat_1,season,model)
                df = pd.read_csv(path+file_sub,index_col=0)
                df.index = pd.to_datetime(df.index)
                all_meanz = []
                for date in df.index:
                    index = int(np.squeeze(np.where(datez_event[:,0]==date)))
                    date_event = datez_event[index]
                    rean_cp_sub = rean_cp.sel(time=slice(date_event[0],
                                                  date_event[1])).mean(dim='time',skipna=True)
                    all_meanz.append(rean_cp_sub)

                fig,ax = plt.subplots(figsize=(16,9))
                rean_cp_sub_all = xr.concat(all_meanz,dim='time').mean(dim='time',skipna=True)
                rean_cp_sub_cyclic, lon_cyclic = addcyclic(rean_cp_sub_all.v.values, rean_cp_sub.longitude)
                m = Basemap(projection='ortho', 
                      lat_0=90, lon_0=0)

        #                 ax=ax)
                m.drawcoastlines()
                latz,lonz = np.meshgrid(rean_cp_sub.latitude,lon_cyclic)
                im = m.contourf(lonz,latz,rean_cp_sub_cyclic.T,
                           latlon='true',cmap='bwr',levels=np.arange(-20,21,1),extend='both')
                fig.suptitle('RWP Phase Speed Composite Analysis {} Forecasts Germany {} {} {}'.format(rank.title(),
                                                                                                 event_titlez[i],
                                                                                                season, model))
                cax = fig.add_axes([0.9, 0.1, 0.05, 0.8])
                ax.text(0.1,0.5,s='N = {}'.format(len(df.index)),transform=fig.transFigure,fontsize=16)
                fig.colorbar(im,orientation='vertical',cax=cax)
                fig.savefig(path_gen+'/fcst_composite/RWP_Phasespeed_Composite_{}_{}_{}_{}'.format(model,event,season,rank),dpi=200)
                plt.close(fig)
```

```python
lat_0,lat_1,lon_0,lon_1 = (54,46,6,14)
file = '{}_forecasts_{}_lon_{}_{}_lat_{}_{}_{}_{}.txt'
seasonz = ['DJF','JJA']
modelz = ['GFS','ERA5RF']

for model in modelz:
    for i,event in enumerate(eventz):
        datez_event = np.load(path_gen + 'GFS_T850/{}_lon_{}_{}_lat_{}_{}.npy'.format(event,lon_0,lon_1-2,lat_0-2,lat_1))
        for season in seasonz:
            for rank in rankz:
                file_sub = file.format(rank,event,lon_0,lon_1,lat_0,lat_1,season,model)
                df = pd.read_csv(path+file_sub,index_col=0)
                df.index = pd.to_datetime(df.index)
                all_meanz_z500 = []
                all_meanz_mslp = []
                for date in df.index:
                    index = int(np.squeeze(np.where(datez_event[:,0]==date)))
                    date_event = datez_event[index]
                    rean_z500_sub = rean_z500.sel(time=slice(date_event[0],
                                                  date_event[1])).mean(dim='time',skipna=True)
                    rean_mslp_sub = rean_mslp.sel(time=slice(date_event[0],
                                                  date_event[1])).mean(dim='time',skipna=True)
                    all_meanz_z500.append(rean_z500_sub)
                    all_meanz_mslp.append(rean_mslp_sub)

                fig,ax = plt.subplots(figsize=(16,9))
                rean_z500_sub_all = xr.concat(all_meanz_z500,dim='time').mean(dim='time',skipna=True)
                rean_z500_sub_cyclic, lon_cyclic = addcyclic(rean_z500_sub_all.z.values, rean_z500_sub.longitude)
                rean_mslp_sub_all = xr.concat(all_meanz_mslp,dim='time').mean(dim='time',skipna=True)
                rean_mslp_sub_cyclic, lon_cyclic = addcyclic(rean_mslp_sub_all.var151.values, rean_mslp_sub.longitude)                
                m = Basemap(projection='ortho', 
                      lat_0=90, lon_0=0)

        #                 ax=ax)
                m.drawcoastlines()
                latz,lonz = np.meshgrid(rean_z500_sub.latitude,lon_cyclic)
                im = m.contourf(lonz,latz,rean_z500_sub_cyclic.T/100,
                           latlon='true',cmap=ncl_colormap(),levels=np.arange(450,601,5),extend='both')
                cl = m.contour(lonz,latz,rean_mslp_sub_cyclic.T/100,colors='k',
                           latlon='true',levels=np.arange(910,1091,6))
                ax.clabel(cl,fmt='%3.0f')
                fig.suptitle('RWP 500 hPa Geopotential and MSLP Composite Analysis {} Forecasts Germany {} {} {}'.format(rank.title(),
                                                                                                 event_titlez[i],
                                                                                                season, model))
                cax = fig.add_axes([0.9, 0.1, 0.05, 0.8])
                ax.text(0.1,0.5,s='N = {}'.format(len(df.index)),transform=fig.transFigure,fontsize=16)
                fig.colorbar(im,orientation='vertical',cax=cax)
                fig.savefig(path_gen+'/fcst_composite/RWP_500_hPa_geop_MSLP_Composite_{}_{}_{}_{}'.format(model,event,season,rank),dpi=200)
                plt.close(fig)
```
