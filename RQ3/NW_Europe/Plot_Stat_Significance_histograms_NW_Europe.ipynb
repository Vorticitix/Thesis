RWP properties of MAE event histograms statistical parameters


```python
path = '/media/onno/Algemeen/Thesis/fcst_RWP_properties/NW_Europe/histograms/'
file_standardized = 'standardized_forecast_errors_stat_significance_lead_day_{}.txt'
file_AV = 'AV_forecast_errors_stat_significance_lead_day_{}.txt'
row_labelz = ['PH DJF','PH JJA','PC DJF','PC JJA',
             'SH DJF','SH JJA','SC DJF','SC JJA']
column_labelz = ['$p_{{env}}$','$p_{{cp}}$','$p_{{T850}}$',
                '$μ_{{env\ good}}$','$μ_{{env\ bad}}$','$μ_{{cp\ good}}$',
                '$μ_{{cp\ bad}}$','$μ_{{T850\ good}}$','$μ_{{T850\ bad}}$']
colorz = [(1,'#ff0000'),(0.75,'#ff6600'),(0.5,'#ffff00'),
          (0.3,'#33cc33'),(0,'#006600')]   
cmap = matplotlib.colors.LinearSegmentedColormap.from_list('custom', colorz[::-1], N=1024)

lead_dayz = [3,5]
for lead_day in lead_dayz:
    df_stand = pd.read_csv(path+file_standardized.format(lead_day),index_col=0)
    df_AV = pd.read_csv(path+file_AV.format(lead_day),index_col=0)
    df_stand_GFS = df_stand.iloc[:,:9]
    df_AV_GFS = df_AV.iloc[:,:9]
    fig,axz = plt.subplots(2,1,figsize=(16,9))
    ax1 = axz.flat[0]
    df_GFS_p = df_stand_GFS.iloc[:,:3]
    df_GFS_p_mask = np.full(df_AV_GFS.shape,np.nan)
    df_GFS_p_mask[:,:3] = df_GFS_p.values
    df_GFS_mu = df_AV_GFS.iloc[:,3:]
    df_GFS_mu_mask = np.full(df_AV_GFS.shape,np.nan)
    df_GFS_mu_mask[:,3:] = df_GFS_mu.values
    ax1.imshow(df_GFS_p_mask,aspect='auto',cmap=cmap,alpha=.4,vmin=0,vmax=1)
    ax1.imshow(df_GFS_mu_mask,aspect='auto',cmap='bwr',alpha=.4,vmin=-1,vmax=1)
    ax1.set_yticks(np.arange(8))
    ax1.set_yticklabels(row_labelz)
    ax1.set_xticks(np.arange(9))
    ax1.set_xticklabels(column_labelz,rotation=35)
    for i in np.arange(len(row_labelz)):
        for j in np.arange(len(column_labelz)):
            if j<3:
                ax1.text(j,i,'{:.3f}'.format(df_stand_GFS.values[i,j]),
                       ha='center',va='center',color='k',fontsize=15)
            else:
                ax1.text(j,i,'{:.3f}'.format(df_AV_GFS.values[i,j]),
                       ha='center',va='center',color='k',fontsize=15)
    df_AV_ERA5RF = df_AV.iloc[:,9:]
    df_stand_ERA5RF = df_stand.iloc[:,9:]
    ax2 = axz.flat[1]
    df_ERA5RF_p = df_stand_ERA5RF.iloc[:,:3]
    df_ERA5RF_p_mask = np.full(df_AV_ERA5RF.shape,np.nan)
    df_ERA5RF_p_mask[:,:3] = df_ERA5RF_p.values
    df_ERA5RF_mu = df_AV_ERA5RF.iloc[:,3:]
    df_ERA5RF_mu_mask = np.full(df_AV_ERA5RF.shape,np.nan)
    df_ERA5RF_mu_mask[:,3:] = df_ERA5RF_mu.values
    ax2.imshow(df_ERA5RF_p_mask,aspect='auto',cmap=cmap,alpha=.4,vmin=0,vmax=1)
    ax2.imshow(df_ERA5RF_mu_mask,aspect='auto',cmap='bwr',alpha=.4,vmin=-1,vmax=1)
    ax2.set_yticks(np.arange(8))
    ax2.set_yticklabels(row_labelz)
    ax2.set_xticks(np.arange(9))
    ax2.set_xticklabels(column_labelz,rotation=35)
    for i in np.arange(len(row_labelz)):
        for j in np.arange(len(column_labelz)):
            if j<3:
                if np.isnan(df_stand_ERA5RF.values[i,j]):
                    continue
                else:
                    ax2.text(j,i,'{:.3f}'.format(df_stand_ERA5RF.values[i,j]),
                       ha='center',va='center',color='k',fontsize=15)
            else:
                if np.isnan(df_AV_ERA5RF.values[i,j]):
                    continue
                else:
                    ax2.text(j,i,'{:.3f}'.format(df_AV_ERA5RF.values[i,j]),
                       ha='center',va='center',color='k',fontsize=15)
    fig.suptitle('Statistical Parameters Standardized Mean Forecast Errors During Extreme Events on Lead Day {}'.format(lead_day))
    fig.savefig(path+'forecast_errors_tile_plot_lead_day_{}'.format(lead_day),dpi=200)
```
